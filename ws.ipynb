```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize the WebDriver (headless mode in this case)
options = webdriver.EdgeOptions()
options.add_argument('--headless')  # Make it headless
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
options.add_argument('--mute-audio')

# driver = webdriver.Edge(options=options)
driver = webdriver.Edge()  # Update path to your chromedriver
  # Update path to your chromedriver

def removenewlines(text):
    """Function to clean the text by removing extra newlines and whitespaces"""
    return text.replace('\n', ' ').replace('\r', '').replace('\s+', ' ').strip()

def waitForElementAndAct(driver, selector, actionCallback):
    """Function that waits for an element and then performs an action on it"""
    timeout=3
    while True:
        start_time = time.time()
        try:
            target_element = driver.find_element(By.CSS_SELECTOR, selector)
            if target_element:
                actionCallback(target_element)
                break
        except:
            pass
        elapsed_time = time.time() - start_time
        if elapsed_time > timeout:
            print(f"Element not found after {timeout} seconds")
            break      # If the element is not found, keep waiting
        time.sleep(0.5)  # Sleep for a while befs
def get_the_transcript(driver):



    #Step 0:Reject Cookies 
    waitForElementAndAct(driver, '#content > div.body.style-scope.ytd-consent-bump-v2-lightbox > div.eom-buttons.style-scope.ytd-consent-bump-v2-lightbox > div:nth-child(1) > ytd-button-renderer:nth-child(1) > yt-button-shape > button',lambda elm: elm.click())
    print("Rejected cookies")
    # Step 1: Wait for and click the "more" button
    waitForElementAndAct(driver, 'tp-yt-paper-button#expand', lambda elm: elm.click())
    print("Clicked the 'more' button")

    # Step 2: Wait for and click the "show transcript" button
    waitForElementAndAct(driver, '#primary-button ytd-button-renderer yt-button-shape button', lambda elm: elm.click())
    print("Clicked 'Show transcript'")

    # Step 3: Wait for the transcript content to load
    waitForElementAndAct(driver, '#segments-container', lambda elm: print("Transcript loaded"))
    
    # Step 4: Extract the transcript text
    transcript_element = driver.find_element(By.CSS_SELECTOR, '#segments-container')
    transcript_text = transcript_element.text
    print("Transcript extracted")

    # Clean up the text by removing unnecessary newlines and extra spaces
    cleaned_transcript = removenewlines(transcript_text)
    print("Transcript cleaned")
    
    return cleaned_transcript

# p = '0iqz4tcKN58'
# Open the YouTube video page
# driver.get('https://www.youtube.com/watch?v=0iqz4tcKN58')  # Replace with the actual video URL

# # Wait for the page to load and collect the transcript
# # time.sleep(3)  # Wait for the page to load (you can adjust this depending on your connection speed)
# transcript = get_the_transcript(driver)

# # Print the cleaned transcript
# print("Cleaned Transcript:")
# print(transcript)

# # Close the browser window
# driver.quit()
```
